```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
dataset = convert(ds)
dataset.head(100)
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
```

#1. Serum Insulin Level Distribution
This graph will show the distribution of serum insulin levels in the dataset, which is crucial because insulin regulation is a central factor in diabetes management.

```python
diabetes_data = pd.read_csv('Diabetes.csv')
plt.figure(figsize=(10, 6))
sns.histplot(diabetes_data['Serum Ins'], kde=True, color='green')
plt.title('Distribution of Serum Insulin Levels')
plt.xlabel('Serum Insulin')
plt.ylabel('Frequency')
plt.show()




```

#2. Diastolic Blood Pressure and Diabetes Class:
The boxplot will compare the diastolic blood pressure readings between diabetic and non-diabetic individuals, exploring how this variable correlates with diabetes outcomes.







```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='Diastolic BP', data=diabetes_data, palette='light:#5A9')
plt.title('Diastolic Blood Pressure by Diabetes Class')
plt.xlabel('Diabetes Class')
plt.ylabel('Diastolic BP')
plt.show()
```

#3. Pair Plot for Insulin, Glucose, and Blood Pressure:
The pair plot will help us visualize the interactions between insulin levels, glucose concentration, and blood pressure, providing a multidimensional view of the dataset.

```python
sns.pairplot(diabetes_data[['PG Concentration', 'Serum Ins', 'Diastolic BP', 'class']], hue='class', palette='deep')
plt.suptitle('Interaction Between Insulin, Glucose, and Blood Pressure', y=1.02)
plt.show()
```

#4. Correlation Heatmap for Selected Variables:
This heatmap will display the correlations between plasma glucose, serum insulin levels, diastolic blood pressure, and BMI, indicating potential relationships for further investigation.

```python
correlation_matrix = diabetes_data[['PG Concentration', 'Serum Ins', 'Diastolic BP', 'BMI']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Selected Variables')
plt.show()
```

#5. Serum Insulin Levels vs. PG Concentration:

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PG Concentration', y='Serum Ins', hue='class', data=diabetes_data, palette='viridis')
plt.title('PG Concentration vs. Serum Insulin by Diabetes Class')
plt.xlabel('PG Concentration')
plt.ylabel('Serum Insulin')
plt.legend(title='Diabetes Class')
plt.show()
```
